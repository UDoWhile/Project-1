```python
import numpy as np
import requests

api_key = 'AIzaSyBZyROvx7Z-lPxM_rrQ0oWrjglvEpRx-nw'

def extract_zip_code(floride_hurricane_2017_2021, api_key, output_format="json"):
    zipcodes = []
    
    for i in range(len(floride_hurricane_2017_2021)):
        base_url = 'https://maps.googleapis.com/maps/api/geocode/'
        lat = floride_hurricane_2017_2021['lat'][i]
        long = floride_hurricane_2017_2021['long'][i]
        endpoint = f'{output_format}?latlng={lat},{long}&key={api_key}'
        
        response = requests.get(base_url + endpoint)
        
        if response.status_code == 200:
            data = response.json()
            
            if 'results' in data and len(data['results']) > 0:
                address_components = data['results'][0]['address_components']
                zip_code = None
                
                for component in address_components:
                    if 'postal_code' in component['types']:
                        zip_code = component['long_name']
                        break
                
                if zip_code is not None:
                    zipcodes.append(zip_code)
                else:
                    zipcodes.append(np.nan)
            else:
                zipcodes.append(np.nan)
        else:
            zipcodes.append(np.nan)
    
    floride_hurricane_2017_2021["zipcodes"] = zipcodes
    return floride_hurricane_2017_2021
```
